```python
import io
 
file = 'BX-Book-Ratings.csv'
with open(file, errors='ignore') as fr:
    data = fr.read()
ratings = pd.read_csv(io.StringIO(data), sep=';', error_bad_lines=False)
```

```python
ratings.head()
```

```python
ratings.rename(columns={'User-ID':'user_id','Book-Rating':'rating'},inplace=True)
ratings.head(2)
```

```python
df.shape
```

```python
users.shape
```

```python
ratings.shape
```

```python
ratings['user_id'].value_counts()
```

```python
ratings['user_id'].value_counts().shape
```

```python
x=ratings['user_id'].value_counts()>200
y=x[x].index
y
```

```python
ratings=ratings[ratings['user_id'].isin(y)]
ratings.shape
```

```python
ratings.head()
```

```python
ratings_with_books=ratings.merge(df,on='ISBN')
ratings_with_books
```

```python
number_rating=ratings_with_books.groupby('title')['rating'].count().reset_index()
number_rating.rename(columns={'rating':'number of ratings'},inplace=True)
number_rating
```

```python
final_ratings=ratings_with_books.merge(number_rating,on='title')
final_ratings.shape
```

```python
final_ratings=final_ratings[final_ratings['number of ratings']>=50]
final_ratings.shape
```

```python
final_ratings.drop_duplicates(['user_id','title'],inplace=True)
final_ratings.shape
```

```python
book_pivot=final_ratings.pivot_table(columns='user_id',index='title',values='rating')
book_pivot
book_pivot.fillna(0,inplace=True)
book_pivot
```

```python
from scipy.sparse import csr_matrix
book_sparse=csr_matrix(book_pivot)
type(book_sparse)
```

```python
from sklearn.neighbors import NearestNeighbors
model=NearestNeighbors(algorithm='brute')
model.fit(book_sparse)
```

```python
distance, suggestions=model.kneighbors(book_pivot.iloc[237, :].values.reshape(1,-1),n_neighbors=6)
```

```python
suggestions
```

```python
distance
```

```python
for i in range(len(suggestions)):
  print(book_pivot.index[suggestions[i]])
```

```python
def recommend_book(book_name):
  book_id=np.where(book_pivot.index==book_name)[0][0]
  distance, suggestions=model.kneighbors(book_pivot.iloc[book_id, :].values.reshape(1,-1),n_neighbors=6)

  for i in range(len(suggestions)):
    if i==0:
      print("The suggestions for",book_name,"are :")
    if not i:
      print(book_pivot.index[suggestions[i]])
```

```python
recommend_book('Animal Farm')
```
